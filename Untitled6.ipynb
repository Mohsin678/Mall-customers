```python
scaler = MinMaxScaler()
```

```python
df["Annual Income (k$)"] = scaler.fit_transform(df[["Annual Income (k$)"]])
```

```python
df["Spending Score (1-100)"] = scaler.fit_transform(df[["Spending Score (1-100)"]])
```

```python
km = KMeans(n_clusters=5)
```

```python
y_pred = km.fit_predict(df[["Annual Income (k$)","Spending Score (1-100)"]])
```

```python
df["cluster1"] =y_pred
```

```python
df.drop("cluster", axis=1, inplace=True)
```

```python
df.head()
```

```python
#build model using scaled data
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster1==0]
df2 = df[df.cluster1==1]
df3 = df[df.cluster1==2]
df4 = df[df.cluster1==3]
df5 = df[df.cluster1==4]

plt.scatter(df1["Annual Income (k$)"], df1["Spending Score (1-100)"], color = "blue")
plt.scatter(df2["Annual Income (k$)"], df2["Spending Score (1-100)"], color = "red")
plt.scatter(df3["Annual Income (k$)"], df3["Spending Score (1-100)"], color = "black")
plt.scatter(df4["Annual Income (k$)"], df4["Spending Score (1-100)"], color = "green")
plt.scatter(df5["Annual Income (k$)"], df5["Spending Score (1-100)"], color = "orange")

plt.xlabel("Annual income")
plt.ylabel("spending score")
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(df[["Annual Income (k$)","Spending Score (1-100)"]])
    sse.append(km.inertia_)
```

```python
sse
```

```python
#check elbow method
```

```python
plt.plot(k_rng,sse)
plt.xlabel("k")
plt.ylabel("sse")
```
